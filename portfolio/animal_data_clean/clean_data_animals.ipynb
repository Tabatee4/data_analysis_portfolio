```python
import pandas as pd
```

```python
data = pd.read_csv('dirty_data_animals.csv',
           sep = ';')
```

```python
display(data)
data_copy = data.copy()
```

```python
data.head()
```

```python
data.info()
```

```python
data.drop(['Animal code'], axis =1, inplace=True)
```

```python
data[data.duplicated()]
```

```python
data.drop_duplicates(inplace=True)
```

```python
data
```

```python
data['Animal type'].unique()
```

```python
data['Animal type'] = data['Animal type'].replace({
    'European bison™': 'European bison',
    'European bisson': 'European bison',
    'European buster': 'European bison',
    'lynx?': 'lynx',
    'red squirel': 'red squirrel',
    'red squirell': 'red squirrel',
    'wedgehod': 'hedgehog',
    'ledgehod': 'hedgehog'
})
data['Animal type'] = data['Animal type'].str.strip().str.title()
data['Animal name'] = data['Animal name'].str.strip().str.title()
```

```python
data
```

```python
data = data.dropna(subset=['Animal type'])
```

```python
data.isna().sum()
```

```python
data = data.copy()

data['Gender'] = data['Gender'].replace('not determined', 'unknown')
data.loc[:, 'Gender'] = data['Gender'].fillna('unknown')
data.loc[:, 'Animal name'] = data['Animal name'].fillna('unknown')
```

```python
data
```

```python
data['Weight kg'] = data['Weight kg'].round(2)
data['Longitude'] = data['Longitude'].round(2)
data['Latitude'] = data['Latitude'].round(2)
```

```python
data.describe()
```

```python
numeric_cols = data.select_dtypes(include='number').columns
data[numeric_cols] = data[numeric_cols].abs()
```

```python
data.dtypes
```

```python
mask_dot = data['Observation date'].str.contains(r'\.', na=False)
data.loc[mask_dot, 'Observation date'] = pd.to_datetime(data.loc[mask_dot, 'Observation date'], format='%d.%m.%Y', errors='coerce')

mask_text = ~mask_dot
data.loc[mask_text, 'Observation date'] = pd.to_datetime(data.loc[mask_text, 'Observation date'], errors='coerce', dayfirst=True)
```

```python
data['Observation date'] = pd.to_datetime(data['Observation date'],format='%Y-%m-%d',errors='coerce')
```
